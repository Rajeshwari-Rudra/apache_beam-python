## Key Concepts:
1. Reading data from text file
2. Specifying 'inline' transforms
3. Counting a PCollection
4. Writing data to text file

* Import the libraries which are required for this project.
  - Imported apache_beam and regular expression libraries as beam and re respectively.
  - Define the P collection pipeline.
  - Reads the data from input file to pipeline.
  - Splits each word of the line from given input file and then add those words to pipeline
  - Using Map function data gets shuffeled and sorted.
  - Sorted data is grouped and returns the sum of word count.
  - CombinePerKey function returns the sum of grouped words.
  - Format_result function formats the result of word count into P collection.
  - Finally, transforms the result from P collection to a textfile.

```
import apache_beam as beam # importing apache_beam and then alias it as beam
import re # importing regular expression

# defining the function format_result for counting each word
def format_result(word_count):
      (word, count) = word_count
      return '%s: %s' % (word, count)

p = beam.Pipeline() #defining P collection pipelines
minimal_word_count=(
    p
    | beam.io.ReadFromText("key_benefits.csv") # reads the input from given text
    | beam.FlatMap(lambda x: re.findall(r'[A-Za-z\']+', x)) # finds the word in each line
    | 'PairWithOne' >> beam.Map(lambda x: (x, 1)) # pairs each word and count it as 1 using Map function
    | 'GroupAndSum' >> beam.CombinePerKey(sum) # Grouping and summing the result
    | 'Format' >> beam.Map(format_result) # formatting the result
    | beam.io.WriteToText("output.txt") # writes the output to the text
)

p.run()
```
