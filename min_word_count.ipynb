```
import apache_beam as beam # importing apache_beam and then alias it as beam
import re # importing regular expression
p1 = beam.Pipeline() # defining a pipeline as p1 and initializing it  
min_word_count = (
    p1
    | beam.io.ReadFromText('comedy_of_errors.txt') # reads the input from file 
    | beam.FlatMap(lambda line: re.findall(r"[a-zA-Z]+",line)) # finds the words 
    | beam.Map(lambda word: (word,1)) # pairing the words with 1
    | beam.CombinePerKey(sum) # Group and sum
    | beam.Map(lambda word_count:str(word_count)) # formatting the results
    | beam.io.WriteToText("output.txt") # writes the output to file
    
)
p1.run()
```
