**Introduction**

netfilx_titles.csv is the data set about the netflix TV showes and movies in which year they are released and on what channels did they telecast and what are they duration.

So on this dataset we do few transformations and look the results in results.txt file.

**Transformations**

Firstly, we filter the netflix TV shows and Movies released on 2020 year and which are only telecasted in 'TV-MA'. On this output we apply the Groupby transformation where we can see all the Movies together as one group and all the TV shows together as other group.

**Program Description**

1. Import the apache beam and alias them as beam
2. Import csv we are using the csv file
3. Create a pipeline and name it as p1
4. Inside pipepline read the netflix_tites.csv, for reoving the header of the file use the skip_header_lines= 1
5. Map the dataset to records which it forms as list.
6. Filter the movies and Tv-shows from data with release year equal to 2020 and telecasted TV channel as TV- MA
7. Now group movies and Tv-shows seperately.
8. Write the output to results.txt file
9. Now run the p1 pipeline.


```
import apache_beam as beam
import csv
p1 = beam.Pipeline()
netflix =(
    p1
    | beam.io.ReadFromText('netflix_titles.csv',skip_header_lines=1) # read the input file and removing the header for the file
    | beam.Map(lambda record: record.split(',')) # create the records for data
    | beam.Filter(lambda record:record[3]=='TV-MA' and (int(record[2])==2020)) # filter the data
    | beam.GroupBy(lambda record:record[1] ) # groupby the data by type record in dataset
    | beam.io.WriteToText('results.txt') # output to results.txt
)
p1.run()
```
