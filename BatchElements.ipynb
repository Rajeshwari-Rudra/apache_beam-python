```
import apache_beam as beam # importing apache_beam and then alias it as beam
import re # importing regular expression

p1 = beam.Pipeline() #defining P collection pipelines
games=(
    p1
    | beam.io.ReadFromText("Games.csv") # reads the input from given text
    | beam.BatchElements(min_batch_size=1,max_batch_size=100)
    | beam.io.WriteToText("output.txt") # writes the output to the text
)

p1.run()
```
